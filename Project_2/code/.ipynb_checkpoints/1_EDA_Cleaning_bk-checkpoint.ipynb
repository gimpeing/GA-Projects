#### Numeric features with strong correlation to target - SalePrice
By going through the sorted list of strong correlation features in previous cell, here are some comments:

a) `Overall Qual` has the highest correlationship with `SalePrice`.

b) `Gr Liv Area` is strongly correlated to `SalePrice`. However, it also strongly correlated to `TotRms AbvGrd` (0.813) and `Full Bath`(0.617), which (`TotRms AbvGrd`, & `Full Bath`) are also in the strong correlation list with `SalePrice`. This means, multicollinearity.

c) `Garage Cars` and `Garage Area` are the next in the list. However, they are both strongly correlated as well (0.893). It makes sense as they are actualy describing the same thing, where `arage area`is the size in square feet and `garage cars` is the size in car capacity. Thus, should keep only one of them, which, `Garage Cars` is selected (as it is slightly higher in correlation to SalePrice).

d) `Year built` is strongly correlated to `SalePrice` However, it also strongly correlated with `Overall Qual`, `Garage Yr Blt`, `Year Remod/Add`. Will take a closer look on these timeline related features before deciding on how to treat them or drop them if found redundency.

e) `Total Bsmt SF` & `1st Flr SF` are both strongly correlated to `SalePrice`. They both also strongly correlated among themselves. Similarly, to keep the `Total Bsmt SF` as `bigger Bsmt SF` will resulted in bigger `1st Flr SF`.

f) `Fireplaces` is somehow seems have value added to `SalePrice`. Same goes to `Mas Vnr Area` and `BsmtFin SF 1`.

g) `Kitchen AbvGr`, `Enclosed Porch`, `PID` gave slightly negative correlation.

```python
# Define function to plot the data and regression model fits across a FacetGrid

def sns_limplot (df, x_factor, y_factor, title, xlabel, ylabel):
    sns.set_style("ticks")
    
    sns.lmplot(x = x_factor,       # Select df column name for x axis element
                y = y_factor,      # Select df column name for y axis element
                data = df,         # Select dataframe
                fit_reg = True,    # Fix a regression line
                scatter_kws={"s": 20},
                height = 6,        # set the height of the plot figure
                aspect = 1.2);     # set the width of the figure, width = aspect*height
    
    plt.title(title, fontsize = 18);
    plt.xlabel(xlabel, fontsize = 14);
    plt.ylabel(ylabel, fontsize = 14);
```

```python
# Define function to plot boxplot for categorical feature

def boxplot_(df, x_col, y_col, xlabel, ylabel, title):
    plt.figure(figsize = (15,12))
    
    # boxplot using seaborn
    ax = sns.boxplot(x=x_col, y=y_col, data=df);
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # Calculate number of occurance (counts) per category & median to position the counts in boxplot
    medians = df.groupby([x_col])[y_col].median().values
    counts = df[x_col].value_counts().values     # count the number of occurance for each category
    counts = [str(x) for x in counts.tolist()]
    counts = ["n: " + i for i in counts]
 
    # Add it to the plot
    pos = range(len(counts))
    for tick,label in zip(pos,ax.get_xticklabels()):
        ax.text(pos[tick], medians[tick] + 0.03, counts[tick],
            horizontalalignment='center', size='x-small', color='k', weight='semibold')
```

## 2. Data Cleaning

Data cleaning on both train dataset and test dataset.
Thus, first step is to combine (concatenate) both of them for data cleaning and split them again.

```python
# added 'data_origin' for df_train & df_tes to differentiate both of them (to ease splitting in later part)
df_train['data_origin'] = 'train'
df_test['data_origin'] = 'test'
```

```python
# combine both datasets by stacking them vertically

df = pd.concat([df_train, df_test], ignore_index = True, sort = False)

# check the shape to ensure they are stack vertically
print(df_train.shape, df_test.shape, df.shape)

df.head()
```

### 2.1 Outliers

#### Gr Liv Area

- In the EDA, it was observed that there are two outliers with large size (sq_ft > 5000) but sold at extreme price. Instead of droping them from the dataset, will change it to the mean value instead.

```python
# Recap on the scatter plot between Gr Liv Area & SalePrice, which shows the outliers

sns_limplot(df_train, 'Gr Liv Area', 'SalePrice', 'Living Area Versus SalePrice',
                'size in sq_ft', 'log(SalePrice), USD')
```

```python
df['Gr Liv Area']
```

```python
# Check if Gr Liv Area > 5000 has only two points, i.e. the outlier mentioned above:

GrLiv_big = df['Gr Liv Area'] > 5000
GrLiv_mean = df['Gr Liv Area'].mean()
df[GrLiv_big].loc[:, ['Gr Liv Area', 'SalePrice']]
```

```python
# Change the size of the outliers to the mean size
df['Gr Liv Area'] = df['Gr Liv Area'].map(lambda x: GrLiv_mean if x > 5000 else x)

# Check have they changed
df[GrLiv_big].loc[:, ['Gr Liv Area', 'SalePrice']]
```

#### Garage Yr Blt

- Another outlier observed in EDA is the Garage Yr Blt is way in future year (year 2207), but was sold in 2007.
- Thus, it should be a error in this data point and decided to change it to same as year remod/add instead.

```python
# Below shows the timeline data for the outlier.
# Construction start in year 2006, remodeling/addtion in year 2007 and sold in the same year. 
# Thus, I believe the garage yr built is highly in year 2007, not 200 year later.

future_garageyrblt = df['Garage Yr Blt'] == 2207.0
df[future_garageyrblt].loc[:,['Garage Yr Blt', 'Year Built', 'Year Remod/Add', 'Yr Sold']]
```

```python
# Replace the outliser of garage yr blt with 2007.
df['Garage Yr Blt'] = df['Garage Yr Blt'].replace({2207.0 : 2007})
```

### 2.2.1 Imputing value `None`

With reference to features description, a number of the features with missing value (or NA, NaN) is correctly recorded. For example, Alley (shown below). 
The NA means no alley access.
   
   - Grvl:  Gravel
   - Pave:  Paved
   - NA:    No alley access

Thus, will impute missing value or NA to None on Alley and other feature that with similar situation.

```python
# list out all the features similar to Alley, that is NA is one of the ordinal rating.

features_w_NA = ['Alley','Mas Vnr Type','Bsmt Qual', 'Bsmt Cond','Bsmt Exposure','BsmtFin Type 1', 
                 'BsmtFin Type 2','Fireplace Qu','Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond',
                'Pool QC', 'Fence', 'Misc Feature']
```

```python
# Impute 'None' to those features identified above
df.update(df[features_w_NA].fillna("None"))
```

```python
# Below are the remaining list with missing value that need to work on
no_value2 = df.isnull().sum()
no_value2 [no_value2>0]
```

### 2.2.2 Imputing value `zero`

To find out the reason for missing value on `Bsmt Full Bath`, `Bsmt Half Bath`,`BsmtFin SF 1`, `BsmtFin SF 2`, `Bsmt Unf SF`, `Total Bsmt SF`
It was observed that the NaN in those columns are due to the house has no basement.
Thus, it will be okay to impute zero on these columns.

```python
# find row with 'Bsmt Full Bath' with NaN.
# it was observed that within the same row,
# 'BsmtFin SF 2','BsmtFin Type 1', 'BsmtFin Type 1' are also NaN, due to no basement
df[df['Bsmt Full Bath'].isnull()].loc[:, ['BsmtFin SF 1','BsmtFin SF 2','BsmtFin Type 1', 'BsmtFin Type 1',
                                       'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']]
```

```python
features_w_NaN1 = ['BsmtFin SF 1','BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF',
                   'Bsmt Full Bath', 'Bsmt Half Bath' ]
```

```python
df.update(df[features_w_NaN1].fillna(0))
```

```python
# find row with 'Garage Cars' with NaN
# Similarly happens to Garage Cars and Garage Area, where NaN is due to no Garage
df[df['Garage Cars'].isnull()].loc[:,['Garage Cars', 'Garage Area', 'Garage Finish', 'Garage Qual']]
```

```python
df.update(df[['Garage Cars', 'Garage Area']].fillna(0))
```

```python
# Find rows with 'Mas Vnr Area' with NaN
# Similarly, it is due to no Masonry venner, (Masonry veneer type = None). Thus, impute with '0'

df[df['Mas Vnr Area'].isnull()].loc[:, ['Mas Vnr Area', 'Mas Vnr Type']]
```

```python
df.update(df[['Mas Vnr Area']].fillna(0))
```

### 2.2.3 Imputing value for a cause

To find out 'Garage Yr Blt' is missing. It was observed that those without Garage Yr Blt is due to there is no garage in these houses.
However, impute with value 'zero' might not appropriate on the the columns with timeseries.
Instead, I will re-assign it to be the same as 'Year Built

```python
# Find rows with 'Garage Yr Blt' with NaN
df[df['Garage Yr Blt'].isnull()].loc[:, ['Year Built', 'Year Remod/Add', 'Garage Yr Blt', 
                                         'Yr Sold', 'Garage Finish']]
```

```python
# To check are all the 159 rows listed above is with no garage (i.e. None in column 'Garage Finish')
df[df['Garage Yr Blt'].isnull()].loc[:,'Garage Finish'].unique()
```

```python
df.update(df['Garage Yr Blt'].fillna(df['Year Built']))
```

### 2.3 Combine features

```python
# Combine Total basement area and Living area above ground
df['Liv_Bsmt_SF'] = df['Gr Liv Area'] + df['Total Bsmt SF']
```

```python
# Combine Total full bathrooms
df['Total_full_bath'] = df['Full Bath'] + df['Bsmt Full Bath']
```

```python
# Combine Total Half bathrooms
df['Total_half_bath'] = df['Half Bath'] + df['Bsmt Half Bath']
```

```python
# Remodel Garage Yr Blt reference to 'Year Built' with conditions:
# +1 if Garage built after Year Built
# 0 if Garage built in year built
# -1 if Garage built before year built

df['garage_yrblt_rel'] = df['Garage Yr Blt'] - df['Year Built']    

df['garage_yrblt_rel'] = df['garage_yrblt_rel'].apply(lambda x: 1 if x >0 else x)
df['garage_yrblt_rel'] = df['garage_yrblt_rel'].apply(lambda x: -1 if x <0 else x)
    
```

```python
# Remodel 'Year Remod/Add' reference to 'Year Built' with conditions:
# +1 if there is remodel/ addition
# 0 if no remodel/ addition

df['remod_yr_rel'] = df['Year Remod/Add'] - df['Year Built'] 

df['remod_yr_rel'] = df['remod_yr_rel'].apply(lambda x: 1 if x >0 else 0)
```
